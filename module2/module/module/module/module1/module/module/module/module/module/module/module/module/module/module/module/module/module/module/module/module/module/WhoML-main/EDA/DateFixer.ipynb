```python
import pandas as pd

df = pd.read_csv('../data/WhoDataSet.csv', parse_dates=['date'])
df.set_index('episode_id', inplace=True)
df.head()
```

```python
df['Weekday'] = df['date'].dt.day_name()
df.head()
```

```python
df[df['Weekday'] == 'Monday'][['title','Producer','rating']].sort_values(by='rating', ascending=False).head(6)
```

```python
df.to_csv('../data/WhoDataSetCategorical.csv')
```
